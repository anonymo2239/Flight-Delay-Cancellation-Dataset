```python
df[df.select_dtypes('object').columns] = df.select_dtypes('object').astype('category')
df.info()
```

```python
semi_clean_df = df.drop(['CANCELLATION_REASON', 'AIR_SYSTEM_DELAY', 'SECURITY_DELAY', 'AIRLINE_DELAY', 'LATE_AIRCRAFT_DELAY', 'WEATHER_DELAY'], axis=1)
semi_clean_df = semi_clean_df.drop(semi_clean_df.loc[(semi_clean_df['DIVERTED'] == 1) | (semi_clean_df['CANCELLED'] == 1)].index, axis=0)
semi_clean_df = semi_clean_df.drop(['YEAR', 'DIVERTED', 'CANCELLED', 'WHEELS_OFF', 'WHEELS_ON', 'DEPARTURE_TIME'], axis=1)
semi_clean_df
```

* Up to this point, semi_clean_df has been cleaned from "unused features" in the model, and object variables have been converted to category.  
* Missing or outlier data has NOT been cleaned yet.  
* The year variable is assigned to "year" since it may be needed for visualization.

### Exploratory Data Analysis (EDA) Findings - My Discoveries:
* The airline company with the most delays is NK, followed by F9.  
* The airline AS and DL almost has no delays.  
* Delays are clearly more frequent on Mondays. 
* Delays are clearly less frequent on Saturdays. 
* The most delays are in June and February. The least delays are in September and October.
* There is no noticeable change in delays as distance increases.  
* TAXI IN and TAXI OUT variables have a clear negative correlation.
* It is noticeable that the departure time is frequent between 06.00 and 10.30 o'clock.
* It is noticeable that the arrival time is frequent between 05.00 and 16.30 o'clock.
